```python
def interploated_y (x,x1,x2,y1,y2):
    y = y1+ ((y2-y1)/(x2-x1))*(x-x1)
    return print(f"Interpolate value: {y} ,x={x},x1={x1},x2={x2},y1={y1},y2={y2}")
```

```python
interploated_y(50,40,60,61.6,71.2)
```

```python
a=(0,20,40,60,100)
for i in a:
    print(i)
print("\n")
b=(26.0,48.6,61.6,71.2,74.8,75.2)
for i in b:
    print(i)
```

```python
import matplotlib.pyplot as plt
```

```python
# Data points
a = (0, 20, 40, 60,80, 100)
b = (26.0, 48.6, 61.6, 71.2, 74.8, 75.2)

# Plotting the graph
plt.plot(a, b, marker='o', linestyle='-', color='b')  # 'o' for dots at data points, '-' for line
plt.xlabel('a (Independent Variable)')  # Label for x-axis
plt.ylabel('b (Dependent Variable)')    # Label for y-axis
plt.title('Graph of b versus a')        # Title of the graph
plt.grid(True)                          # Show grid lines for better readability

# Display the graph
plt.show()
```
