```python
x=[3,4,5,6,7,8]
y=[0,7,17,26,35,45]
n=len(x)
sumx=sumxy=sumx2=sumy=0  
for i in range(n):
    sumx+=x[i]
    sumxy+=x[i]*y[i]
    sumx2+=x[i]**2
    sumy+=y[i]
print(f"sum of x = {sumx},sum of xi*yi = {sumxy},sum of x2 = {sumx2},sum of yi = {sumy}")
xmean=sumx/n
ymean=sumy/n
a=(ymean*sumx2-xmean*sumxy)/(sumx2-n*xmean**2)
b=(sumxy-xmean*sumy)/(sumx2-n*xmean**2)
print(f'xmean={xmean},ymean={ymean},a={a},b={b}')
print(f"Equation of straightline:y={a}x+{b}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,100,100)
print(x)
y=a*x+b
print('\n',y)
```

```python
plt.plot(x,y,label=f'y={a}x+{b}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')
plt.grid(True)
plt.legend()
plt.show()
```
