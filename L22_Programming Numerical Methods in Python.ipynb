```python
import numpy as np

# Given data points
x = np.array([0, 1, 2, 3, 4, 5], float)
y = np.array([2, 8, 14, 28, 39, 62], float)

# Number of data points and the degree of polynomial
m = len(x)
n = 2  # 2 for quadratic equation

# Initialize matrix A and vector B
A = np.zeros((n + 1, n + 1))
B = np.zeros(n + 1)

# Calculate the elements of matrix A and vector B
for row in range(n + 1):
    for col in range(n + 1):
        if row == 0 and col == 0:
            A[row, col] = m
        else:
            A[row, col] = np.sum(x**(row + col))
    B[row] = np.sum(x**row * y)

# Solve for the coefficients a0, a1, a2
a = np.linalg.solve(A, B)

# Display the result
print("The result of the polynomial:")
print('f(x) = %f' % a[0])
for i in range(1, n + 1):
    print('\t %+f x^%d' % (a[i], i))
```
