```python
import sympy as sp
```

```python
# Define the variable
x = sp.symbols('x')

# Define the function
f = 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2

# Calculate the derivative with respect to x
f_derivative = sp.diff(f,x)

# Display the derivative
print("First-order derivative:")
f_derivative
```

```python
# Define the variable
x = sp.symbols('x')

# Define the function
f = 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2

# Calculate the first-order derivative with respect to x
f_derivative = sp.diff(f, x)

# Calculate the second-order derivative by differentiating the first-order derivative
f_second_derivative = sp.diff(f_derivative, x)

# Display the second-order derivative
print("Second-order derivative:")
print(f_second_derivative)
```

```python
# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = 0.1*x**5 - 0.2*x**3*y + 0.1*x*y - 0.2*y**2

# Calculate the partial derivative with respect to x
f_partial_x = sp.diff(f, x)

# Calculate the partial derivative with respect to y
f_partial_y = sp.diff(f, y)

# Display the partial derivatives
f_partial_x, f_partial_y
```

```python
# Define the variables
x, y = sp.symbols('x y')

# Define the function
f = 0.1*x**5 - 0.2*x**3*y + 0.1*x*y - 0.2*y**2

# Calculate the partial derivative with respect to x
f_partial_x = sp.diff(f, x)

# Calculate the partial derivative with respect to y
f_partial_y = sp.diff(f, y)

# Define the point at which to evaluate the partial derivatives
x0, y0 = 1, 2  # Example point (x0, y0)

# Substitute the point into the partial derivatives
f_partial_x_at_point = f_partial_x.subs({x: x0, y: y0})
f_partial_y_at_point = f_partial_y.subs({x: x0, y: y0})

# Display the results
print("Partial derivative with respect to x at (x0, y0):")
print(f_partial_x_at_point)

print("\nPartial derivative with respect to y at (x0, y0):")
print(f_partial_y_at_point)
```
