```python
# Define the function to integrate
def f(x, y):
    return x**2 * y + x * y**2  # Given function

# Integration limits and number of intervals
ax = 1  # Lower limit of x
bx = 2  # Upper limit of x
ay = -1  # Lower limit of y
by = 1  # Upper limit of y
nx = 10  # Number of divisions in x direction (must be even)
ny = 10  # Number of divisions in y direction (must be even)

# Calculate step sizes
hx = (bx - ax) / nx  # x step size
hy = (by - ay) / ny  # y step size

# Initialize the summation
S = 0.0

# Apply Simpson's 1/3 Rule in 2D
for i in range(ny + 1):  # Loop for y direction
    y_i = ay + i * hy
    if i == 0 or i == ny:
        p = 1  # Factor for the first and last rows
    elif i % 2 == 1:
        p = 4  # Factor for odd rows
    else:
        p = 2  # Factor for even rows
    
    for j in range(nx + 1):  # Loop for x direction
        x_j = ax + j * hx
        if j == 0 or j == nx:
            q = 1  # Factor for the first and last columns
        elif j % 2 == 1:
            q = 4  # Factor for odd columns
        else:
            q = 2  # Factor for even columns
        
        # Add the weighted function value to the summation
        S += p * q * f(x_j, y_i)

# Apply scaling factor for Simpson's 1/3 Rule
Integral = (hx * hy / 9.0) * S

# Print the result
print(f'Integral = {Integral:.6f}')
print(f'Integral = {Integral}')
```
