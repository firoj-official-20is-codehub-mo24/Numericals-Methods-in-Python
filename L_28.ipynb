```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its analytical derivatives
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x + 0.2
diff_f1 = lambda x: 0.5*x**4 - 0.6*x**2 + 0.1
diff_f2 = lambda x: 2.0*x**3 - 1.2*x

# Set the step size and point of evaluation
h = 0.005
x = np.linspace(-1, 1, 25)  # 25 points for reshaping into 5x5

# Compute the finite differences
dff1 = (f(x + h) - f(x - h)) / (2 * h)  # Central difference for the first derivative
dff2 = (f(x + h) - 2 * f(x) + f(x - h)) / h**2  # Central difference for the second derivative

# Reshape the results into 5x5 matrices
f_matrix = f(x).reshape((5, 5))
dff1_matrix = dff1.reshape((5, 5))
dff2_matrix = dff2.reshape((5, 5))

# Print the results
print("x")
print(x.reshape((5, 5)))
print("f(x) =")
print(f_matrix)
print("\nf'(x) =")
print(dff1_matrix)
print("\nf''(x) =")
print(dff2_matrix)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), '-k', label='f(x)')        # Plot f(x) with a solid black line
plt.plot(x, dff1, '--b', label="f'(x)")      # Plot f'(x) with a dashed blue line
plt.plot(x, dff2, '--r', label="f''(x)")     # Plot f''(x) with a dashed red line
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Function and Its Derivatives")
plt.legend()  # Display the legend
plt.grid()
plt.show()
```
