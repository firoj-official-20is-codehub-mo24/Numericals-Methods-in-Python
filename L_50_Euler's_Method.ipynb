```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential function dy/dx = x * y
dy = lambda x, y: x * y

# Analytical solution y = e^(x^2 / 2)
def analytical_solution(x):
    return np.exp(x**2 / 2)

# Euler's method function
def euler_method(f, x0, y0, x_end, h):
    x = np.arange(x0, x_end + h, h)  # Create an array of x values from x0 to x_end with step size h
    y = np.zeros(len(x))             # Initialize the array for y values
    y[0] = y0                        # Set the initial condition

    # Perform Euler's method
    for i in range(1, len(x)):
        y[i] = y[i - 1] + h * f(x[i - 1], y[i - 1])

    return x, y

# Initial conditions
x0 = 0       # Initial value of x
y0 = 1       # Initial value of y
x_end = 2    # End value of x
h = 0.01     # Step size

# Compute the Euler method approximation
x_euler, y_euler = euler_method(dy, x0, y0, x_end, h)

# Compute the analytical solution
x_analytical = np.arange(x0, x_end + h, h)
y_analytical = analytical_solution(x_analytical)

# Print and plot the results
print("Euler's Method Results:")
for xi, yi in zip(x_euler, y_euler):
    print(f"x:{xi:.1f} and y_euler:{yi:.4f} and y_analytical:{analytical_solution(xi):.4f}")

# Plotting the results
plt.plot(x_euler, y_euler, 'o-', label='Euler\'s Method')
plt.plot(x_analytical, y_analytical, 'r--', label='Analytical Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Euler\'s Method vs Analytical Solution')
plt.legend()
plt.grid(True)
plt.show()
```
